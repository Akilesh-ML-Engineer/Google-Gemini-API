```python
import google.generativeai as genai
import os
import pprint
import textwrap
from IPython.display import Markdown
from IPython.display import display
```

```python
# Setting the api key
API_KEY = os.environ["GOOGLE_GEMINI_API"]
genai.configure(api_key=API_KEY)
```

```python
# Listing all the models
for model in genai.list_models():
    print(f"{model.name}  -   {model.supported_generation_methods}")
```

```python
def to_markdown(plain_text):
  text = plain_text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
model = genai.GenerativeModel('gemini-1.5-pro')
```

```python
model
```

```python
%%time
response = model.generate_content("Explain me about the AGI and whether it is possible possible to attain it ?")
```

```python
to_markdown(response.text)
```

```python
response
```

When using <code>GenerativeModel.generate_content(..., stream=True)</code>, the model outputs response chunks as they are generated, allowing for real-time streaming of the response instead of waiting for the entire generation process to complete.

```python
%time

stream_response = model.generate_content("Explain the stream param in the generate_content() method of the GenerativeModel class of the google_generativeai python package", stream=True)
```

```python
for chunk in stream_response:
    print(chunk.text)
    print("_"*80)
```

### Multimodal input

Multimodal inputs refer to the ability of a machine learning model or system to process and interpret multiple types of input data simultaneously. These inputs can come from various modalities such as text, images, audio, and video. The goal of using multimodal inputs is to enable the model to gain a more comprehensive understanding of the information by integrating different types of data, which often leads to improved performance in tasks like classification, generation, and retrieval.

```python
prompt = "Explain me about the car apperance in the picture"
img1
```

## Chat conversation

```python
model
```

The `ChatSession` class in Gemini keeps track of all the back-and-forth in the conversation automatically. So, you don't need to manually manage or store the history of your conversation. This makes it much easier to have `multi-turn conversations` (where you go back and forth several times) because `ChatSession` remembers what was said earlier and uses that information to generate appropriate responses.

#### Multi-turn models
A multi-turn conversation involves multiple exchanges between a user and AI, with the AI remembering and considering previous parts of the conversation to provide relevant and coherent responses. This makes interactions more natural and contextually aware.

```python
chat_model = model.start_chat(
    history=[]
)
chat_model
```

```python
chat_response_1 = chat_model.send_message("Hi how are you?")
chat_response_1
```

```python
to_markdown(chat_response_1.text)
```

In `ChatSession`, the chat history is stored as a list of `glm.Content` objects, each containing multiple `glm.Part` objects. These parts can either be text or binary data (like images), with the binary data being represented as `glm.Blob` objects that include the data and its MIME type.

```python
img1
```

```python
chat_model.send_message(img1)
```

```python
chat_model.history
```

```python
# To see all the message in the chachat_model
for message in chat_model.history:
    print(message)
    print("-"*80)
```

```python
for message in chat_model.history:
    display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```

```python
# Count tokens
model.count_tokens(chat_model.history)
```
